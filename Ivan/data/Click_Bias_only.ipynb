# Get the ones with the clickbait or bias

```python
import pandas as pd
import numpy as np
import os
import random
import csv
import sys
maxInt = sys.maxsize
while True:
    # decrease the maxInt value by factor 10 
    # as long as the OverflowError occurs.

    try:
        csv.field_size_limit(maxInt)
        break
    except OverflowError:
        maxInt = int(maxInt/10)
```

```python
csv.field_size_limit(maxInt)
csv_iter = pd.read_csv('./news_cleaned_2018_02_13.csv', engine='python', iterator=True, chunksize=50000, encoding='utf-8', header=0, error_bad_lines=False, dtype={'type': str, 'content': str,'authors': str})
print("STARTING GETTING THE ONE WITH BIAS ONLY")
i = 1
for single_df in csv_iter:
    modified_df = single_df[(single_df['type'] == 'bias') | (single_df['type'] == 'clickbait')]
    modified_df.to_csv('./click_bias_only.csv', mode='a', index=False)
    print(f"DONE ITERATION {i}")
    i = i + 1
print("DONE ALL")
```
